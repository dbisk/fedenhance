```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.reload_library()
plt.style.use(['science', 'ieee', 'vibrant'])
```

```python
df = pd.read_csv('snr_data.csv', index_col=0)
```

```python
fig = plt.figure()
fig.set_size_inches(5.22, 2)
ax = sns.violinplot(data=df, y="Splits", x="SNR(dB)", hue="Noises", split=True, inner="quart")

plt.legend(ncol=1, bbox_to_anchor=(1.01, 0.53), frameon=False, shadow=False, fontsize=10, handletextpad=0.2)
plt.xlim([-40, 50])
plt.ylabel('')
plt.xlabel('Input SNR (dB)', fontsize=13)
plt.xticks(fontsize=10)
plt.yticks(fontsize=13)

for l in ax.lines[1::3]:
    l.set_linestyle('-')

plt.savefig('input_snrs_violin.pdf')
plt.show()
```
